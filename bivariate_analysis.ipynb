# PROJECT INDIAN KIDS SCREEN TIME 
# BIVARIATE ANALYSIS

```python
# Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading Dataset
df = pd.read_csv('ScreenSense_Cleaned_Dataset.csv')
df.head()
```

# Screen Time Based Analysis
1. Screen Timr vs Age Bond

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Age_Band", y="Avg_Daily_Screen_Time_hr", palette="Set2")

plt.title("Average Screen Time by Age Band")
plt.xlabel("Age Band")
plt.ylabel("Average Daily Screen Time (Hours)")
plt.show()
```

 In This analysis Screen time increases with age. The 16–20 age group shows the highest average daily screen time compared to younger age

2. Gender vs Screen time

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="Gender",
    y="Avg_Daily_Screen_Time_hr",
    palette="coolwarm"
)

plt.title("Average Screen Time by Gender", fontsize=14, fontweight='bold')
plt.xlabel("Gender")
plt.ylabel("Average Daily Screen Time (Hours)")

plt.tight_layout()
plt.show()
```

In This Analysis Shows Average daily screen time between different genders. Male students demonstrate a higher average daily screen time compared to female students

3. Screen Time vs Location Type 

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x="Urban_or_Rural",
    y="Avg_Daily_Screen_Time_hr",
    ci=None,
    palette="viridis"
)

plt.title("Average Screen Time: Urban vs Rural")
plt.xlabel("Location Type")
plt.ylabel("Average Daily Screen Time (Hours)")
plt.show()
```

Rural students spend marginally more time on screens than urban students, but the difference is very small. This indicates that screen time behavior is becoming uniform across different geographical locations.

# Device Based Analysis  

1. Average Screentime by Device

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="Primary_Device",
    y="Avg_Daily_Screen_Time_hr",
    palette="Set2"  
)

plt.title("Average Screen Time by Primary Device")
plt.xlabel("Primary Device")
plt.ylabel("Average Daily Screen Time (Hours)")
plt.xticks(rotation=20)

plt.show()
```

Students using Laptop show higher average screen time compared to other devices, indicating higher engagement levels.


2. Device usage by ageband

```python
plt.figure()
sns.countplot(
    data=df,
    x="Age_Band",
    hue="Primary_Device"
)
plt.title("Device Usage by Age Band")
plt.xlabel("Age Band")
plt.ylabel("Number of Students")

plt.show()
```

```python
df.groupby("Primary_Device")["Avg_Daily_Screen_Time_hr"].mean()
```

3. device usage by Gender 

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Primary_Device",
    hue="Gender",
    palette="Set2"
)
plt.title("Primary Device Usage by Gender")
plt.xlabel("Primary Device")
plt.ylabel("Number of Students")
plt.xticks(rotation=20)
plt.show()
```

Boys use most devices more than girls, except laptops and Smartphone is the most popular device for both genders.

4,  Activity catogery by Gender

```python
plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    x="Gender",
    hue="Activity_Category",
    palette="Set2"
)
plt.title("Activity Category by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.xticks(rotation=20)
plt.show()
```

Most boys and girls use screens in a balanced way. Boys slightly use screens more for recreational purposes, but the difference between boys and girls is small
