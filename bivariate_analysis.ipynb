# Bivariate Analysis

### Screen Based Analysis

```python
#library import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset
df=pd.read_csv("Screen_Time_Cleaned.csv")
df.head()
```

##### 1. Screen time vs Age band

```python
age_screen_time = df.groupby("Age_Band")["Avg_Daily_Screen_Time_hr"].mean()
print(age_screen_time)

# plt.figure()
# sns.barplot(x=age_screen_time.index, y=age_screen_time.values, palette="Set2")
# plt.title("Average Daily Screen Time by Age Band")
# plt.xlabel("Age Band")
# plt.ylabel("Average Daily Screen Time (hours)")
# plt.show()

sns.boxplot(x="Age_Band", y="Avg_Daily_Screen_Time_hr", data=df, palette="Set2")
```

##### Insight:
           Screen time is almost sililar across all ages but younger childern show slightly more extreme usage in some cases.

##### 2. Gender vs Screen Time

```python
gender_screen= df.groupby("Gender")["Avg_Daily_Screen_Time_hr"].mean()
print(gender_screen)

plt.figure()
sns.barplot(x=gender_screen.index, y=gender_screen.values, palette="Set1")
plt.title("Average Daily Screen Time by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Daily Screen Time (hours)")
plt.show()
```

##### Insight:
          Both male and female spend nearly the same amount of time on screen , showing no major difference by gender.

##### 3. Screen time vs Location

```python
location_screen=df.groupby("Urban_or_Rural")["Avg_Daily_Screen_Time_hr"].mean()
print(location_screen)

plt.figure()
sns.barplot(x=location_screen.index, y=location_screen.values, palette="Set2")
plt.title("Average Daily Screen Time by Location")
plt.xlabel("Location")
plt.ylabel("Average Daily Screen Time (hours)")
plt.show()
```

##### Insight:
    Childern in both urban and rural areas have almost equal screen time, indicating location does not affect usage much.

### Device Based Analysis

##### 1.Device vs Screen Time

```python
device_screen = df.groupby("Primary_Device")["Avg_Daily_Screen_Time_hr"].mean()
print(device_screen)

plt.figure()
sns.barplot(x=device_screen.index, y=device_screen.values, palette="viridis")
plt.title("Average Daily Screen Time by Device Type")
plt.xlabel("Device Type")
plt.ylabel("Average Daily Screen Time (hours)")
plt.show()
```

##### Insight:
     Children using laptops and smartphones tend to spend slightly more time on screens compared to TV  and tablets.

##### 2.Device vs Gender

```python
device_usage=pd.crosstab(df["Primary_Device"], df["Usage_Type"])
device_usage.plot(kind="bar", stacked=True )
plt.title("Usage Type Distribution by Primary Device")
plt.xlabel("Primary Device")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.show()
```

##### Insight:
     Most of the children uses their devices more for recreational purpose than for educational purpose.

```python
plt.figure()
sns.countplot(data=df,x="Age_Band",hue="Primary_Device", palette="Set1")
plt.title("Device Usage by Age Band")
plt.xlabel("Age Band")
plt.ylabel("Number of Students")
plt.show()
```

##### Insight:
     Smartphones are the most commmomly used device in all age groups,especially among childern ah=ged 11-15.
