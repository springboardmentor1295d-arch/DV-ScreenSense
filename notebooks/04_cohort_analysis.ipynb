```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/screensense_clean.csv")

df.head()
```

### Basic Cohort Analysis

```python
# Create age cohort groups
df["Age_Cohort"] = pd.cut(
    df["Age"],
    bins=[5,9,13,17,20],
    labels=["6-9","10-13","14-17","18-20"]
)

# Count users per cohort
cohort = df.groupby("Age_Cohort").size().reset_index(name="Count")

plt.figure(figsize=(8,5))

sns.barplot(
    data=cohort,
    x="Age_Cohort",
    y="Count"
)

plt.title("Basic Cohort Analysis by Age Group")

plt.savefig("../outputs/figures/cohort_analysis.png")

plt.show()
```

Cohort analysis groups users into cohorts based on shared characteristics.
In this case, children are grouped into age cohorts to analyze the distribution
of users across different age groups.
