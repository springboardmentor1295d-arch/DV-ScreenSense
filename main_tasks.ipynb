# 1. Age Band × Device Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('ScreenSense_Cleaned_Dataset.csv')
df.head()
pivot_table = df.pivot_table(
    values="Avg_Daily_Screen_Time_hr",
    index="Age_Band",
    columns="Primary_Device",
    aggfunc="mean"
)
plt.figure(figsize=(8,5))
sns.heatmap(
    pivot_table,
    annot=True,
    cmap="YlOrRd",
    fmt=".2f"
)
plt.title("Average Screen Time by Age Band and Device")
plt.xlabel("Primary Device")
plt.ylabel("Age Band")
plt.show()
```

Older age groups generally show higher screen time across most devices.

Laptop and smartphone usage dominate across multiple age bands, indicating these devices are the primary sources of screen exposure.

# Task 2 — Screen Time vs Health Impact

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df,
    x="Health_Risk_Flag",
    y="Avg_Daily_Screen_Time_hr",
    palette="Set2"
)

plt.title("Average Screen Time by Health Risk")
plt.xlabel("Health Risk")
plt.ylabel("Average Screen Time (Hours)")

plt.savefig("screen_time_health_risk.png")
plt.show()
```

Students with health risks tend to have higher screen time compared to those without reported health issues.

The distribution shows greater variability among high screen-time users, suggesting excessive usage may contribute to health problems.

# Task 3 — Urban vs Rural Screen Time

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x="Urban_or_Rural",
    y="Avg_Daily_Screen_Time_hr",
    ci=None,
    palette="viridis"
)

plt.title("Average Screen Time: Urban vs Rural")
plt.xlabel("Location Type")
plt.ylabel("Average Daily Screen Time (Hours)")
plt.show()
```

Rural students show slightly higher average screen time compared to urban students.

However, the difference is small, suggesting digital device usage is common across both regions.
