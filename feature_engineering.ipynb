Age band

```python
import pandas as pd
df=pd.read_csv("Indian_Kids_Screen_Time.csv")
df["age_band"] = pd.cut(
    df["Age"],
    bins=[7, 9, 13, 17, 18],
    labels=["8–9", "10–13", "14–17", "18"],
    right=True
)
```

```python
df["age_band"].value_counts()
```

Screen Time Buckets

```python
df["screentime_bucket"] = pd.cut(
    df["Avg_Daily_Screen_Time_hr"],
    bins=[0, 2, 5, df["Avg_Daily_Screen_Time_hr"].max()],
    labels=["Low", "Medium", "High"],
    include_lowest=True
)
```

```python
df["screentime_bucket"].value_counts()
```

Usage Balance (Based on educational ratio)

```python
df["Usage_Balance"] = pd.cut(
        df["Educational_to_Recreational_Ratio"],
        bins=[0, 0.4, 0.5, 1],
        labels=["More Recreational", "Balanced", "More Educational"],
        include_lowest=True
    )
```

```python
df["Usage_Balance"].value_counts()
```

Save cleaned dataset

```python
df.to_csv("screensense_clean.csv", index=False)

pd.read_csv("screensense_clean.csv").head()
```

Feature Dictionary

Student gender (categorical).

Average number of hours spent daily on screens.

Boolean indicating whether screen time exceeds recommended daily limit.

Limitations

Health impact data had missing values (~33%), which were assumed as "No Reported Issues".
Dataset does not contain socio-economic or parental supervision factors.
Screen time is self-reported, which may introduce response bias.

Assumptions

Missing health impact entries indicate no reported issues.
Screen time bucket thresholds (2h, 5h) are based on common digital wellness recommendations.
Educational ratio thresholds (0.4, 0.5) are used for logical classification and may vary in real-world studies.
