```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

loading the dataset

```python
df=pd.read_csv('Screen_Time.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna("None", inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df.info()
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df['Gender'].unique()
```

```python
df["Age"].unique()
```

```python
def age_group(age):
    if age <=10:
        return "6-10"
    elif age <=15:
        return "11-15"
    else:
        return "16-18"
    
df["Age_Band"]=df["Age"].apply(age_group)
df["Age_Band"].value_counts()
```

```python
def screen_time_category(screen_time):
    if screen_time <= 3:
        return "Low"
    elif screen_time <= 5:
        return "Medium"
    else:
        return "High"
    
df["Screen_Time_Level"]=df["Avg_Daily_Screen_Time_hr"].apply(screen_time_category)
df["Screen_Time_Level"].value_counts()
```

```python
def health_impact_flag(impact):
    if impact == "None":
        return "No risk"
    else:
        return "At risk"
    
df["Health_Risk_Flag"]=df['Health_Impacts'].apply(health_impact_flag)    
df["Health_Risk_Flag"].value_counts()
```

```python
def usage_type(row):
    if row>=0.5:
        return "More Educational"
    else:
        return "More Recreational"
df["Usage_Type"]=df["Educational_to_Recreational_Ratio"].apply(usage_type)    
df["Usage_Type"].value_counts()
```

```python
df.to_csv('Screen_Time_Cleaned.csv', index=False)
```
