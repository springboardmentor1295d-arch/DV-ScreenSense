DATA CLEANING

```python
#health impacts have missing values we will change it to "None"
df['Health_Impacts']=df['Health_Impacts'].fillna("None")
```

```python
#checking for null values again
df.isnull().sum()
```

Standerdisze categories

```python
#Gender 
df['Gender'].unique()
```

```python
df['Gender']=df['Gender'].str.strip().str.title()
```

```python
#cheking for primary device used by the users
df['Primary_Device'].unique()
```

```python
#removing whitespace and converting to title case
df['Primary_Device']=df['Primary_Device'].str.strip().str.title() 
```

```python
#checking if urban or rural users are more
df['Urban_or_Rural'] = df['Urban_or_Rural'].str.strip().str.title()
```

Removing invalid rows

```python
#age validation
df= df[(df['Age'] >= 5) & (df['Age'] <= 18)]
```

```python
#screen time validation values cant be negative
df = df[(df['Avg_Daily_Screen_Time_hr'] >=0) & (df['Avg_Daily_Screen_Time_hr'] <= 24)]
```

```python
#education to recreation ratio validation
df=df[df['Educational_to_Recreational_Ratio'] >= 0]
```

```python
df=df.drop_duplicates() #removing duplicates
```

Feature engineering

```python
#creating age bands
df['Age_Band'] = pd.cut(df['Age'], bins=[0, 5,9,13, 17, 18], labels=['3-5', '6-9', '10-13', '14-17', '18+'])

df['Age_Band'].value_counts().sort_index()
#df['Age_Band'] = pd.cut(df['Age'], bins=[5, 8, 12, 15, 18], labels=['5-8', '9-12', '13-15', '16-18'])



```

```python
df['Screen_Time_Level'] = pd.cut(
    df['Avg_Daily_Screen_Time_hr'],
    bins=[0, 2, 4, 6, 8, 24],
    labels=['Low', 'Moderate', 'High', 'Very High', 'Extreme'],
)
```
