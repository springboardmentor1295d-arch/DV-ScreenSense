### Outlier Detection in Screen Time Data

In this section, we identify **extreme screen time values (outliers)** in the dataset using the **Interquartile Range (IQR) method**. Outliers are data points that fall far outside the normal range and may represent unusual or excessive screen usage.

First, the dataset is cleaned by removing rows with missing values in the **Health_Impacts** column and standardizing categorical fields such as **Urban_or_Rural** and **Primary_Device**.

The IQR method is then used to calculate the **lower and upper bounds** for normal screen time values. Any observations outside this range are considered outliers.

The analysis includes:
- Calculating **Q1 (25th percentile)** and **Q3 (75th percentile)** of screen time
- Determining the **IQR (Interquartile Range)** and identifying extreme values
- Counting the **total number of outliers**
- Analyzing how these outliers are distributed across **locations (Urban vs Rural)** and **device types**

Finally, **box plots** are used to visually detect outliers in:
- **Urban vs Rural screen time**
- **Device-based screen time usage**

This analysis helps identify **students with unusually high screen usage** and determine whether specific **locations or devices contribute to extreme screen time behavior**.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure cleaned dataset
df_cleaned = df.dropna(subset=["Health_Impacts"])

# Standardize text columns
df_cleaned["Urban_or_Rural"] = df_cleaned["Urban_or_Rural"].str.strip().str.title()
df_cleaned["Primary_Device"] = df_cleaned["Primary_Device"].str.strip().str.title()

# -----------------------------------
# 1️⃣ Detect Outliers using IQR Method
# -----------------------------------

Q1 = df_cleaned["Avg_Daily_Screen_Time_hr"].quantile(0.25)
Q3 = df_cleaned["Avg_Daily_Screen_Time_hr"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Identify outliers
outliers = df_cleaned[
    (df_cleaned["Avg_Daily_Screen_Time_hr"] < lower_bound) |
    (df_cleaned["Avg_Daily_Screen_Time_hr"] > upper_bound)
]

print("\nTotal Outliers Found:", len(outliers))

# -----------------------------------
# 2️⃣ Outliers by Urban/Rural
# -----------------------------------

print("\nOutliers by Location:")
print(outliers["Urban_or_Rural"].value_counts())

# -----------------------------------
# 3️⃣ Outliers by Primary Device
# -----------------------------------

print("\nOutliers by Device:")
print(outliers["Primary_Device"].value_counts())

# -----------------------------------
# 4️⃣ Visualize Outliers (Urban vs Rural)
# -----------------------------------

plt.figure(figsize=(6,5))
sns.boxplot(x="Urban_or_Rural",
            y="Avg_Daily_Screen_Time_hr",
            data=df_cleaned)

plt.title("Outlier Detection: Urban vs Rural")
plt.xlabel("Location")
plt.ylabel("Screen Time (Hours)")
plt.show()

# -----------------------------------
# 5️⃣ Visualize Outliers (Device Based)
# -----------------------------------

plt.figure(figsize=(6,5))
sns.boxplot(x="Primary_Device",
            y="Avg_Daily_Screen_Time_hr",
            data=df_cleaned)

plt.title("Outlier Detection: Device-Based Screen Time")
plt.xlabel("Primary Device")
plt.ylabel("Screen Time (Hours)")
plt.xticks(rotation=45)
plt.show()
```

### Health Impact vs Screen Time by Age Band

This analysis explores the relationship between **age groups, health impacts, and average daily screen time**.

First, the dataset is cleaned by removing rows with missing values. Then an **Age_Band** column is created to group students into age ranges (3–5, 6–9, 10–13, and 14–17).

A **pivot table** is created to calculate the **average screen time for each combination of age band and reported health impact**.

Finally, the results are visualized using a **heatmap**, where:
- Rows represent **Age Bands**
- Columns represent **Health Impact categories**
- Colors represent the **average screen time values**

This visualization helps identify patterns between **screen usage and health-related issues across different age groups**.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure dataset exists
df_cleaned = df.dropna()

# Create Age Bands
df_cleaned["Age_Band"] = pd.cut(
    df_cleaned["Age"],
    bins=[0,5,9,13,17],
    labels=["3-5","6-9","10-13","14-17"]
)

# Create pivot table
health_age = pd.pivot_table(
    df_cleaned,
    values="Avg_Daily_Screen_Time_hr",
    index="Age_Band",
    columns="Health_Impacts",
    aggfunc="mean"
)

# Plot heatmap
plt.figure(figsize=(8,6))

sns.heatmap(
    health_age,
    annot=True,
    cmap="coolwarm"
)

plt.title("Health Impact vs Screen Time by Age Band")
plt.xlabel("Health Impact")
plt.ylabel("Age Band")

plt.show()
```

### Screen Time Limit Exceedance by Age Group

In this step, we analyze which **age groups exceed the recommended daily screen time limit**.

First, a new column called **Exceeded_Limit** is created.  
This column checks whether a student's **average daily screen time is greater than 2 hours**, which is considered the recommended limit.

### Creating Health Risk Flag

In this step, a new column called **Health_Risk_Flag** is created to categorize students based on their **average daily screen time**.

The `pd.cut()` function is used to divide screen time into three risk levels:
- **Low Risk** → Screen time ≤ 2 hours  
- **Moderate Risk** → Screen time between 2 and 5 hours  
- **High Risk** → Screen time > 5 hours  

This classification helps identify **students at different levels of screen usage risk**.

Finally, a sample of the dataset is displayed to verify that the **Health_Risk_Flag** column has been created correctly.

```python
# Create Health Risk Flag column
df_cleaned["Health_Risk_Flag"] = pd.cut(
    df_cleaned["Avg_Daily_Screen_Time_hr"],
    bins=[0, 2, 5, float("inf")],
    labels=["Low Risk", "Moderate Risk", "High Risk"]
)

# Check result
print(df_cleaned[["Avg_Daily_Screen_Time_hr", "Health_Risk_Flag"]].head())
```
