# MULTIVARIATE ANALYSIS

Age Band × Device × Screen Time

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("ScreenSense_Cleaned_Dataset.csv")

plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="Age_Band",
    y="Avg_Daily_Screen_Time_hr",
    hue="Primary_Device",
    palette="Set2"
)

plt.title("Screen Time by Age Band and Device")
plt.xlabel("Age Band")
plt.ylabel("Average Screen Time (Hours)")

plt.show()
```

• Screen time generally increases with age, especially in the 16–20 age group.
• Smartphones and laptops contribute the highest screen time across most age groups.

Gender × Device × Screen Time

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="Primary_Device",
    y="Avg_Daily_Screen_Time_hr",
    hue="Gender",
    palette="coolwarm"
)

plt.title("Screen Time by Device and Gender")
plt.xlabel("Primary Device")
plt.ylabel("Average Screen Time (Hours)")

plt.show()
```

• Male students show slightly higher screen time across most devices.

• Laptop and smartphone usage show the highest screen time for both genders.

Age Band × Activity Category × Screen Time

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="Age_Band",
    y="Avg_Daily_Screen_Time_hr",
    hue="Activity_Category",
    palette="viridis"
)

plt.title("Screen Time by Age Band and Activity Type")
plt.xlabel("Age Band")
plt.ylabel("Average Screen Time (Hours)")

plt.show()
```

• Students with balanced activity patterns tend to have moderate screen time.

• Older students show more recreational screen usage compared to younger children.
