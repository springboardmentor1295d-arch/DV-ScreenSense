## INDIAN KIDS SCREEN TIME ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading Dataset
df = pd.read_csv('Preprocessed_Data.csv')
```

### UNIVARIATE ANALYSIS

##### 1. Average Daily Screen Time Distribution

```python
plt.figure(figsize=(7,4))
plt.hist(df['Avg_Daily_Screen_Time_hr'],bins=10,color='skyblue',edgecolor='black')
plt.title("Distribution of Average Daily Screen Time")
plt.xlabel("Average Daily Screen Time (Hours)")
plt.ylabel("Number of Students")
plt.show()
```


Mostly children spend 4-5 hours daily, whereas few use screen excessively, leading to risky concern

##### 2. Primary Device Distribution

```python
plt.figure(figsize=(8,5))
device_counts = df["Primary_Device"].value_counts()
plt.bar(device_counts.index, device_counts.values,color='lightgreen', 
        edgecolor='black', linewidth=0.5)
plt.title("Primary Device Distribution")
plt.xlabel("Primary Device")
plt.ylabel("Children Count")
plt.xticks(rotation=20)
# Add values on bars
for i, v in enumerate(device_counts.values):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()
```

Smartphone are the most commonly used devices by children,
indicating that children rely more on portable and easily accessible technology.

##### 3. Gender Distribution

```python
plt.figure(figsize=(6,4))
gender_counts = df["Gender"].value_counts()
bars = plt.bar(gender_counts.index, gender_counts.values,
               color=['lightblue', 'lightpink'], edgecolor='black', linewidth=0.5)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Children")
plt.bar_label(bars)
plt.show()
```

The dataset shows a fairly balanced distribution between male and female students. 

##### 4. Location Distribution

```python
plt.figure(figsize=(6,4))
loc_counts=df["Urban_or_Rural"].value_counts()
bars=plt.bar(loc_counts.index,loc_counts.values,color='mistyrose',edgecolor='black',linewidth=0.5)
plt.title("Urban vs Rural Distribution")
plt.xlabel("Location")
plt.ylabel("Count")
plt.bar_label(bars,padding=3)
plt.show()
```

Urban students are slightly more represented in the dataset,
This may indicate higher digital access in urban areas.

##### 5. Health Impact Distribution

```python
plt.figure()
health_counts=df["Health_Impacts"].value_counts().sort_values()
bars=plt.barh(health_counts.index,health_counts.values,color='brown',edgecolor='black',linewidth=0.5)
plt.title("Health Impact Distribution")
plt.xlabel("Count")
plt.ylabel("Health Impact")
plt.bar_label(bars,padding=3)
plt.show()
```

Several students report health issues like eye strain or sleep problems

##### 6. Screen Time Level Distribution

```python
plt.figure()
level_counts=df["Screen_Time_Level"].value_counts()
bars=plt.bar(level_counts.index,level_counts.values,color='peachpuff',edgecolor='black',linewidth=0.5)
plt.title("Screen Time Level Distribution")
plt.xlabel("Screen Time Level")
plt.ylabel("Count")
plt.bar_label(bars,padding=2)
plt.show()
```

Most students fall under medium to high screen time levels
This indicates increasing screen usage among students.

### BIVARIATE ANALYSIS

### Screen Time Based Analysis

##### 1. Average Screen Time by Age Band

```python
plt.figure(figsize=(7,4))
avg_screen=df.groupby('Age_Band')['Avg_Daily_Screen_Time_hr'].mean().sort_index()
bars=plt.bar(avg_screen.index,avg_screen.values,color='blue',edgecolor='black',linewidth=0.5)
plt.title("Average Screen Time by Age Band")
plt.xlabel("Age Band")
plt.ylabel("Avg Screen Time (Hours)")
plt.xticks(rotation=0)
plt.bar_label(bars, fmt='%.2f')
plt.show()
```

The trend shows a clear increase in screen usage with age, indicating that older students are more engaged with digital devices

##### 2. Average Screen Time by Location(Urban/Rural)

```python
plt.figure(figsize=(6,4))
df.groupby('Urban_or_Rural')['Avg_Daily_Screen_Time_hr'].mean().plot(
    kind='bar', color=['green', 'red'], edgecolor='black',linewidth=0.5)
plt.title("Urban vs Rural Screen Time")
plt.xlabel("Location")
plt.ylabel("Avg Screen Time (Hours)")
plt.xticks(rotation=0)
plt.bar_label(bars,fmt='%.2f')
plt.show()
```

Despite having low data, Rural children tend to show higher screen time

##### 3. Average Screen Time by Gender

```python
plt.figure(figsize=(6,4))
data=df.groupby("Gender")["Avg_Daily_Screen_Time_hr"].mean()
bars=plt.bar(data.index,data.values,color=['pink','lightblue'],edgecolor='black',linewidth=0.5)
plt.title("Average Screen Time by Gender")
plt.xlabel("Gender")
plt.ylabel("Avg Screen Time (hrs)")
plt.bar_label(bars,fmt='%.1f',padding=3)
plt.show()
```

There is slight difference between both genders in average screen time consumption

##### 4. Average Screen Time by Health Impact

```python
plt.figure(figsize=(8,5))
data=df.groupby("Health_Impacts")["Avg_Daily_Screen_Time_hr"].mean().sort_values()
bars=plt.barh(data.index,data.values,color='peachpuff',edgecolor='black',linewidth=0.5)
plt.title("Avg Screen Time by Health Impact")
plt.xlabel("Avg Screen Time")
plt.ylabel("Health Impact")
plt.bar_label(bars,fmt='%.1f',padding=3)
plt.show()
```

Eye strain and sleep issues are the most common health impacts due to increased screen time. Low screen time students mostly fall under No Issues category

##### 5. Screen Time Level vs Health Risk

```python
plt.figure(figsize=(8,5))
data=df.groupby("Screen_Time_Level")["Health_Risk_Flag"].value_counts().unstack()
ax=data.plot(kind='bar',color=['lightgreen','lightcoral'],edgecolor='black',linewidth=0.5)
plt.title("Screen Time Level vs Health Risk")
plt.xlabel("Screen Time Level")
plt.ylabel("Count")
plt.xticks(rotation=0)
for container in ax.containers:
    ax.bar_label(container,padding=3)
plt.show()
```

### Device Based Analysis

##### 1. Average Screen Time by Primary Device

```python
plt.figure(figsize=(7,4))
data=df.groupby("Primary_Device")["Avg_Daily_Screen_Time_hr"].mean()
bars=plt.bar(data.index,data.values,color='orange',edgecolor='black',linewidth=0.5)
plt.title("Avg Screen Time by Primary Device")
plt.xlabel("Primary Device")
plt.ylabel("Avg Screen Time")
plt.xticks(rotation=20)
plt.bar_label(bars,fmt='%.2f')
plt.show()
```

All devices nearly show high usage, laptops and smartphones are slightly used more

##### 2. Health Risk by Device

```python
plt.figure(figsize=(8,5))
data=df.groupby("Primary_Device")["Health_Risk_Flag"].value_counts().unstack()
ax=data.plot(kind='bar',color=['lightgreen','lightcoral'],edgecolor='black',linewidth=0.5)
plt.title("Device vs Health Risk")
plt.xlabel("Device")
plt.ylabel("Count")
plt.xticks(rotation=20)
for container in ax.containers:
    ax.bar_label(container,padding=3)
plt.show()
```

Students using mobile devices show higher health risks.

##### 3. Screen Time Level by Device

```python
plt.figure(figsize=(8,5))
data=df.groupby("Primary_Device")["Screen_Time_Level"].value_counts().unstack()
ax=data.plot(kind='bar',edgecolor='black',linewidth=0.5)
plt.title("Device vs Screen Time Level")
plt.xlabel("Device")
plt.ylabel("Count")
plt.xticks(rotation=0)
for container in ax.containers:
    ax.bar_label(container,padding=3)
plt.show()
```

Mobile users are mostly in medium and high screen time categories.
Other device users are more balanced across levels.

##### 4. Percentage Exceeding Recommended Limit by Device

```python
plt.figure(figsize=(7,4))
data=df.groupby("Primary_Device")["Screen_Time_Level"].apply(lambda x:(x=="High").mean()*100)
bars=plt.bar(data.index,data.values,color='lightblue',edgecolor='black',linewidth=0.5)
plt.title("Percentage Exceeding Screen Time Limit by Device")
plt.xlabel("Device")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=20)
plt.bar_label(bars,fmt='%.1f',padding=3)
plt.show()
```

Mobile users have the highest percentage exceeding recommended screen time

##### 5. Device Usage Balance by Age Band

```python
plt.figure(figsize=(8,5))
data=df.groupby("Age_Band")["Primary_Device"].value_counts().unstack()
ax=data.plot(kind='bar',edgecolor='black',linewidth=0.5)
plt.title("Usage Balance by Age Band")
plt.xlabel("Age Band")
plt.ylabel("Count")
plt.xticks(rotation=0)
for container in ax.containers:
    ax.bar_label(container,padding=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
age_device = pd.crosstab(df['Age_Band'], df['Primary_Device'])
age_device.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5),
    colormap='viridis'
)
plt.title("Age Band vs Primary Device Usage")
plt.xlabel("Age Band")
plt.ylabel("Number of Students")
plt.xticks(rotation=0)
plt.legend(title="Device Type")
plt.show()
```

Younger and older age groups show different device preferences.

### MULTIVARIATE ANALYSIS

##### 1. Age Band + Primary Device + Screen Time Analysis

```python
multi1 = df.groupby(['Age_Band', 'Primary_Device'])['Avg_Daily_Screen_Time_hr'].mean().unstack()
multi1.plot(
    kind='bar',
    figsize=(9,5),
    colormap='tab20'
)
plt.title("Avg Screen Time by Age Band & Device")
plt.xlabel("Age Band")
plt.ylabel("Avg Screen Time (Hours)")
plt.xticks(rotation=0)
plt.legend(title="Device Type")
plt.show()
```

Screen time is influenced by both age and device type, with certain device-age combinations showing significantly higher usage.
Screen time generally increases with age, especially in the 16–18 age group. 
Smartphones and laptops contribute the highest screen time across most age groups.

##### 2. Age Band + Location + Screen Time Analysis

```python
plt.figure(figsize=(8,5))
data=df.groupby(["Age_Band","Urban_or_Rural"])["Avg_Daily_Screen_Time_hr"].mean().unstack()
bars=data.plot(kind='bar',color=['lightgreen','lightblue'],edgecolor='black',linewidth=0.5)
plt.title("Age Band & Location vs Avg Screen Time")
plt.xlabel("Age Band")
plt.ylabel("Avg Screen Time (hrs)")
plt.xticks(rotation=0)
for container in bars.containers:
    bars.bar_label(container,fmt='%.1f',padding=3)
plt.show()
```

##### 3. Device + Screen Time Level → Health Risk

```python
from matplotlib.colors import LinearSegmentedColormap
plt.figure(figsize=(8,5))
data=df[df["Health_Risk_Flag"]=="Risk"].groupby(["Primary_Device","Screen_Time_Level"]).size().unstack()
colors=["lightyellow","orange","red"]
cmap=LinearSegmentedColormap.from_list("risk_map",colors)
plt.imshow(data,aspect='auto',cmap=cmap)
plt.colorbar(label="Risk Count")
plt.xticks(range(len(data.columns)),data.columns)
plt.yticks(range(len(data.index)),data.index)

# Add values
for i in range(len(data.index)):
    for j in range(len(data.columns)):
        val=data.iloc[i,j] if not pd.isna(data.iloc[i,j]) else 0
        plt.text(j,i,int(val),ha='center',va='center')

plt.title("Device + Screen Time Level → Health Risk")
plt.xlabel("Screen Time Level")
plt.ylabel("Device")
plt.show()
```

Health risk is highest for high screen time users, especially on mobile devices.”

##### 4. Age Band × Primary Device Heatmap

```python
plt.figure(figsize=(7,5))
data=pd.crosstab(df["Age_Band"],df["Primary_Device"])

plt.imshow(data,cmap='Reds',aspect='auto')
plt.colorbar(label="Count")

plt.xticks(range(len(data.columns)),data.columns)
plt.yticks(range(len(data.index)),data.index)

# Add values inside cells
for i in range(len(data.index)):
    for j in range(len(data.columns)):
        plt.text(j,i,data.iloc[i,j],ha='center',va='center')

plt.title("Age Band vs Primary Device")
plt.xlabel("Primary Device")
plt.ylabel("Age Band")

plt.show()
```

The heatmap highlights that device preference varies slightly with age, but mobile usage dominates across all age bands.

### COHORT ANALYSIS

### Exceeded Limit Analysis

```python
df["Exceeded_Limit"]=df["Screen_Time_Level"].apply(lambda x:"Yes" if x=="High" else "No")
```

##### 1. Age Band vs Exceeded Recommended Limit

```python
plt.figure(figsize=(7,4))
data=df.groupby("Age_Band")["Exceeded_Limit"].value_counts().unstack()
ax=data.plot(kind='bar',color=['lightgreen','lightcoral'],edgecolor='black',linewidth=0.5)
plt.title("Age Band vs Exceeded Screen Time Limit")
plt.xlabel("Age Band")
plt.ylabel("Count")
plt.xticks(rotation=0)
for container in ax.containers:
    ax.bar_label(container,padding=3)
plt.show()
```

Older age groups exceed screen time limits more frequently.

##### 2. Device vs Exceeded Recommended Limit

```python
plt.figure(figsize=(7,4))
data=df.groupby("Primary_Device")["Exceeded_Limit"].value_counts().unstack()
ax=data.plot(kind='bar',color=['lightgreen','lightcoral'],edgecolor='black',linewidth=0.5)
plt.title("Device vs Exceeded Screen Time Limit")
plt.xlabel("Device")
plt.ylabel("Count")
plt.xticks(rotation=20)
for container in ax.containers:
    ax.bar_label(container,padding=3)
plt.show()
```

Mobile users exceed recommended screen time limits the most.


##### 3. Location vs Exceeded Limit

```python
plt.figure(figsize=(6,4))
data=df.groupby("Urban_or_Rural")["Exceeded_Limit"].value_counts().unstack()
ax=data.plot(kind='bar',color=['lightgreen','lightcoral'],edgecolor='black',linewidth=0.5)
plt.title("Location vs Exceeded Screen Time Limit")
plt.xlabel("Location")
plt.ylabel("Count")
for container in ax.containers:
    ax.bar_label(container,padding=3)
plt.xticks(rotation=0)
plt.show()
```

Urban students exceed screen time limits more than rural students.

##### 4. Top Health Impact vs Exceeded Limit

```python
top=df["Health_Impacts"].value_counts().head(5).index
data=df[df["Health_Impacts"].isin(top)]
data=data.groupby("Health_Impacts")["Exceeded_Limit"].value_counts().unstack().fillna(0)
bars=data.plot(kind='barh',edgecolor='black',linewidth=0.5)
plt.title("Top Health Impacts vs Exceeded Limit")
plt.xlabel("Count")
plt.ylabel("Health Impact")
for container in bars.containers:
    bars.bar_label(container,padding=3)
plt.show()
```

Students exceeding screen time limits show higher counts in major health issues.

Overall, Exceeding recommended screen time is more common among older, urban, and mobile users, and is strongly linked to increased health issues.
