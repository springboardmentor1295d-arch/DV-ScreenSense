# INDIAN KIDS SCREEN TIME

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Indian_Kids_Screen_Time.csv")
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

#### Handle Missing Values

```python
df["Health_Impacts"] = df["Health_Impacts"].fillna("No Issues")
```

```python
df['Health_Impacts']
```

#### Clean and Standardize Text

```python
# remove spaces + standardize To ensure consistency in categorical values
#  for accurate grouping and visualization
df["Gender"] = df["Gender"].str.strip().str.capitalize()
df["Urban_or_Rural"] = df["Urban_or_Rural"].str.strip().str.capitalize()
df["Primary_Device"] = df["Primary_Device"].str.strip().str.title()
df["Exceeded_Recommended_Limit"] = df["Exceeded_Recommended_Limit"].astype(str).str.strip().str.capitalize()
```

#### Check for Invalid Values

```python
# Age range check
df = df[(df["Age"] >= 8) & (df["Age"] <= 18)]

# Ratio check
df = df[(df["Educational_to_Recreational_Ratio"] >= 0) & 
        (df["Educational_to_Recreational_Ratio"] <= 1)]

# Avg_Daily_Screen_Time_hr
df = df[(df['Avg_Daily_Screen_Time_hr'] >= 0) & 
        (df['Avg_Daily_Screen_Time_hr'] <= 24)]
```

#### Remove Duplicates

```python
df = df.drop_duplicates()
```

```python
df.shape
```

#### Feature Engineering

```python
df['Age_Band'] = pd.cut(df['Age'], bins=[0, 10, 15, 18], labels=['8-10', '11-15', '16-18'])
```

```python
df['Screen_Time_Level'] = pd.cut(
    df['Avg_Daily_Screen_Time_hr'],
    bins=[0, 2, 4, 6, 24],
    labels=['Low', 'Moderate', 'High', 'Very High'],
    include_lowest=True
)
```

```python
df["Health_Risk_Flag"] = df["Health_Impacts"].apply(
    lambda x: "No Risk" if x == "No Issues" else "Risk")
```

#### Final Check

```python
df.shape
df.isnull().sum()
df.info
```

#### Convert into Preprocessed_Data file

```python
df.to_csv('Preprocessed_Data.csv', index=False)
```
