```python
#plot1- Age vs Avg screentime
pivot=pd.pivot_table(
    data=df,
    values="Avg_Daily_Screen_Time_hr",
    index="Age",
    aggfunc="mean"
)
print(pivot)
plt.bar(pivot.index,pivot["Avg_Daily_Screen_Time_hr"])
```

```python
plt.plot(pivot.index,pivot["Avg_Daily_Screen_Time_hr"])
plt.xlabel("Age")
plt.ylabel("Avg_Daily_Screen_Time_hr")
plt.title("Age vs Avg screentime")
plt.show()
```

```python
#plot2- age vs gender vs avg screentime
pivot=pd.pivot_table(
    data=df,
    columns="Gender",
    values="Avg_Daily_Screen_Time_hr",
    index="Age",
    aggfunc="mean"
)
print(pivot)
```

```python
plt.figure()
plt.plot(pivot.index,pivot["Female"],color="red")
plt.plot(pivot.index,pivot["Male"])
plt.xlabel("Age")
plt.ylabel("Avg_Daily_Screen_Time_hr")
plt.title("Age vs Gender vs Avg screentime")
plt.legend(["Female","Male"])
plt.show()
```

```python
#count of avg_screentime_by_hr by age and gender
pivot=pd.pivot_table(
    data=df,
    columns="Gender",
    values="Avg_Daily_Screen_Time_hr",
    index="Age",
    aggfunc="count"
)
print(pivot)
```

```python
plt.figure()
plt.plot(pivot.index,pivot["Female"])
plt.plot(pivot.index,pivot["Male"])
plt.xlabel("Age")
plt.ylabel("Avg_Daily_Screen_Time_hr")
plt.title("Age vs Gender vs Avg screentime")
plt.legend(["Female","Male"])
plt.show()
```

```python
#plot4- Educational to recreational ratio vs age
plt.figure()
plt.scatter(df['Educational_to_Recreational_Ratio'],df['Age'])
plt.xlabel("Educational_to_Recreational_Ratio")
plt.ylabel("Avg_Daily_Screen_Time_hr")
```

```python
plt.figure()
plt.scatter(df['Educational_to_Recreational_Ratio'],df['Avg_Daily_Screen_Time_hr'])
plt.xlabel("Educational_to_Recreational_Ratio")
plt.ylabel("Avg_Daily_Screen_Time_hr")
plt.title("Educational_to_Recreational_Ratio vs Avg_Daily_Screen_Time_hr")
plt.show()
```

```python
from os import device_encoding
#plot5 - DISTRIBUTION ACROSS PRIMARY DEVICES
device_counts=df["Primary_Device"].value_counts()
plt.figure(figsize=(8,5))
plt.bar(device_counts.index,device_counts.values)
plt.xlabel("Primary_Device")
plt.ylabel("No of kids")
plt.title("DISTRIBUTION ACROSS PRIMARY DEVICES")
plt.show()
```

```python
#plot5 - average screentime vs device
avg_screentime = df.groupby('Primary_Device')['Avg_Daily_Screen_Time_hr'].mean()

# Plot bar chart
plt.figure(figsize=(8, 5))
avg_screentime.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Average Screen Time by Device')
plt.xlabel('Device')
plt.ylabel('Average Screen Time (hours)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# plot - avg screentime > 4 hr by gender for laptop users
print(df.columns)

filtered_df = df.query("`Avg_Daily_Screen_Time_hr` > 4 and Primary_Device == 'Laptop'")
avg_screentime_gender = filtered_df.groupby('Gender')['Avg_Daily_Screen_Time_hr'].mean()

# Plot bar chart
plt.figure(figsize=(8, 5))
avg_screentime_gender.plot(kind='bar', color='lightgreen', edgecolor='black')

plt.title('Average Screen Time (>4 hrs) by Gender for Laptop Users')
plt.xlabel('Gender')
plt.ylabel('Average Screen Time (hours)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

```python
#Distribution of daily screen time
plt.hist(df['Avg_Daily_Screen_Time_hr'],color='green')
plt.xlabel("Avg_Daily_Screen_Time_hr")
plt.ylabel("Frequency")
plt.title("Distribution of daily screen time by hr")
plt.show()
```

```python
#plot 6- comparision between avg screentime and location
location_avg=df.groupby("Urban_or_Rural")["Avg_Daily_Screen_Time_hr"].mean()
plt.figure(figsize=(8,5))
plt.bar(location_avg.index,location_avg.values)
plt.xlabel("Urban_or_Rural")
plt.ylabel("Avg_Daily_Screen_Time_hr")
plt.title("comparision between avg screentime and location")
plt.show()
```

```python
df["Age_Band"] = pd.cut(
    df["Age"],
    bins=[0, 5, 9, 13, 17],
    labels=["3–5", "6–9", "10–13", "14–17"]
)
```

```python
df["Screen_Time_Level"] = pd.cut(
    df["Avg_Daily_Screen_Time_hr"],
    bins=[0, 2, 4, 6, 24],
    labels=["Low", "Moderate", "High", "Very High"]
)
```

```python
df["Age_Band"] = pd.cut(
    df["Age"],
    bins=[5, 9, 13, 17],
    labels=["6–9", "10–13", "14–17"]
)
age_band_mean = (
    df.groupby("Age_Band")["Avg_Daily_Screen_Time_hr"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(8, 5))
sns.barplot(
    data=age_band_mean,
    x="Age_Band",
    y="Avg_Daily_Screen_Time_hr"
)

plt.title("Average Daily Screen Time Increases with Age")
plt.xlabel("Age Band")
plt.ylabel("Average Daily Screen Time (Hours)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(
    data=df,
    x="Age_Band",
    y="Avg_Daily_Screen_Time_hr"
)

plt.title("Screen Time Variability Across Age Bands")
plt.xlabel("Age Band")
plt.ylabel("Average Daily Screen Time (Hours)")
plt.show()
```

```python
# Calculate average screen time by gender
avg_screen_time = df.groupby('Gender')["Avg_Daily_Screen_Time_hr"].mean()

print("Average Screen Time by Gender:")
print(avg_screen_time)

# Plot
avg_screen_time.plot(kind='bar', color=['pink', 'blue'])
plt.title('Average Screen Time vs Gender')
plt.ylabel('Average Screen Time (hours)')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.show()
```

```python
# Calculate average age by gender
avg_age = df.groupby('Gender')['Age'].mean()

print("Average Age by Gender:")
print(avg_age)

# Plot
avg_age.plot(kind='bar', color=['pink', 'blue'])
plt.title('Average Age vs Gender')
plt.ylabel('Average Age (years)')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.show()
```

```python
# Calculate average screen time by health impact
avg_screen_time = df.groupby('Health_Impacts')['Avg_Daily_Screen_Time_hr'].mean()

print("Average Screen Time by Health Impact:")
print(avg_screen_time)

# Plot
avg_screen_time.plot(kind='bar', color=['blue', 'yellow', 'pink'],width=1)
plt.title('Average Screen Time vs Health Impact')
plt.ylabel('Average Screen Time (hours)')
plt.xlabel('Health Impact')
plt.figure(figsize=(1, 1))
plt.xticks(rotation=0)
plt.show()
```

```python
impact_counts = df.groupby(['Gender', 'Health_Impacts']).size().unstack(fill_value=0)

print("Health Impact Counts by Gender:")
print(impact_counts)

# Plot stacked bar chart
impact_counts.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Gender vs Health Impacts')
plt.ylabel('Count')
plt.xlabel('Gender')
plt.xticks(rotation=0)

plt.legend(title='Health Impact')
plt.show()
```
