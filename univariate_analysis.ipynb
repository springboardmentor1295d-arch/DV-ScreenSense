# Univariate Analysis

```python
#library import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loding the dataset
df=pd.read_csv("Screen_Time_Cleaned.csv")
```

#### 1. Screen time Distribution

```python
plt.Figure()
sns.histplot(df["Avg_Daily_Screen_Time_hr"],bins=15, kde=True)
plt.title("Distribution of Average Daily Screen Time") 
plt.xlabel("Average Daily Screen Time (hours)")
plt.ylabel("Frequency")
plt.show()
```

##### Insight:
              Screen time is  concentrated around 4-6 hours per day, indicating moderate usage among most children.
              

#### 2. Age Band Distribution

```python
age_counts = df["Age_Band"].value_counts()

plt.figure()
sns.barplot(x=age_counts.index, y=age_counts.values, palette="Set2")
```

##### Insight:
            The dataset is dominated by the 11-15 age group, with nearly double the observations compared to other groups.

#### 3. Screen Time Level

```python
level_counts = df["Screen_Time_Level"].value_counts()

plt.figure()
sns.barplot(x=level_counts.index, y=level_counts.values, palette="viridis")
plt.title("Screen Time Level Distribution")
plt.show()
```

##### Insight:
        Majority of the children fall under Medium and High Screen time level, indicating widespread moderate to heavy usage.

#### 4. Primary Device Distribution

```python
device_counts = df["Primary_Device"].value_counts()

plt.figure()
sns.barplot(x=device_counts.index, y=device_counts.values, palette="Set1")
plt.title("Primary Device Distribution")
plt.xticks(rotation=30)
plt.xlabel("Primary Device")
plt.ylabel("Count")
plt.show()
```

##### Insight:
              Smartphones are the most commonly used among children.
