# Indian Kids Screen Time Analysis
# Univariate analysis

```python
# library import
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# loading Dataset
df = pd.read_csv('ScreenSense_Cleaned_Dataset.csv')
```

1. Screen Time

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Avg_Daily_Screen_Time_hr"], bins=10, kde=True)

plt.title("Distribution of Average Daily Screen Time")
plt.xlabel("Average Daily Screen Time (Hours)")
plt.ylabel("Frequency")

plt.show()
```

Most children spend around 4–5 hours daily on screens. While extreme usage is not common, a small number of students exceed 8 hours, which may raise health and behavioral concerns.

2. Primary Device

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x="Primary_Device", palette="Set2")
plt.title("Primary Device Distribution")
plt.xticks(rotation=20)
plt.show()
```

Smartphone is the most commonly used primary device among students.
Tablets are the least used device.
